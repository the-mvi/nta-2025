## Основные теоретические положения


### Основные определения

#### Социоматрица
Социоматрица $A$ - это квадратная матрица размера $g \times g$, где $g$ - количество акторов в сети. Элемент $a_{ij}$ принимает значение:
- $1$, если существует связь от актора $i$ к актору $j$
- $0$, если связь отсутствует

Для неориентированных сетей матрица симметрична: $a_{ij} = a_{ji}$

#### Список рёбер
Альтернативное представление сети в виде множества упорядоченных пар:
$E = \{(i,j) | a_{ij} = 1\}$

#### Плотность сети
Плотность неориентированной сети G вычисляется как:

$$\Delta = \frac{2L}{g(g-1)}$$

где $L$ - количество рёбер, $g$ - количество вершин.

#### Компоненты связности
Компонента связности - максимальное подмножество вершин, где существует путь между любой парой вершин.

#### Диаметр сети
Диаметр $diam$ - максимальное геодезическое расстояние между любой парой вершин в сети G:

$$diam(G) = \max_{i,j} d_{ij}$$

где $d_{ij}$ - длина кратчайшего пути между вершинами $i$ и $j$.

#### Коэффициент кластеризации
Локальный коэффициент кластеризации $Cl_i$ для вершины $i$ графа G:

$$Cl_i(G) = \frac{2N_i}{k_i(k_i-1)}$$

где $N_i$ - количество треугольников через вершину $i$, $k_i$ - степень вершины.

Глобальный коэффициент кластеризации:

$$Cl(G) = \frac{1}{n} \sum_{i=1}^n C_i$$

#### Двудольные сети

Сеть $G = (V, E)$, где $V = V_1 \cup V_2$, $V_1 \cap V_2 = \emptyset$, и все рёбра $e \in E$ соединяют вершины из разных множеств:
$$
E \subseteq \{(u,v) | u \in V_1, v \in V_2\}
$$


#### Гиперграф

Обобщение графа, в котором каждым ребером могут соединяться не два узла, а подмножество

$$
H = H(A, B)$$ где

$A$ - набор вершин

$B$ - набор гипер ребер


#### Проекции двудольной сети

- *Проекция на акторы* ($P_1$): $A_1 = B B^T - D_1$
- *Проекция на события* ($P_2$): $A_2 = B^T B - D_2$

где $D_1$, $D_2$ - диагональные матрицы степеней.


#### K-ядра

$k$-ядро - максимальный подграф, в котором каждый узел смежен не менее чем с k узлами в подграфе:

$$\deg(v) \geq k, ∀ v \in V$$


#### Отношение шансов

Для таблицы сопряженности:

| | Участвует в $j$ | Не участвует в $j$ |
|-|-----------------|-------------------|
| Участвует в $i$ | $a$ | $b$ |
| Не участвует в $i$ | $c$ | $d$ |

$$
OR = \frac{a \cdot d}{b \cdot c}, \quad LOR = \ln(OR)
$$





### Меры центральности

#### Центральность по степени
Для вершины $i$ в неориентированной сети:

$$C_D(i) = deg(i)$$

Нормированная центральность по степени:

$$C_D'(i) = \frac{deg(i)}{g - 1}$$

#### Центральность по близости
$$C_C(i) = \frac{1}{\sum_{j \neq i} d_{ij}}$$

где $d_{ij}$ - расстояние между вершинами.

Нормированная центральность по близости:

$$C_C'(i) = \frac{g-1}{\sum_{j \neq i} d_{ij}}$$

#### Центральность по посредничеству
$$C_B(i) = \sum_{j < k} \frac{\sigma_{jk}(i)}{\sigma_{jk}}$$

где $\sigma_{jk}$ - количество кратчайших путей между $j$ и $k$, $\sigma_{jk}(i)$ - количество таких путей через $i$.

Нормированная центральность по посредничеству:

$$C_B'(i) =\frac{2 C_B(i)}{(g-1)(g-2)}$$

#### Центральность по информации

Алгоритм нахождения:
1) Вычислить матрицу псевдосмежности $A = 1 + K$, где $K$ - матрица Кирхгофа $g \times g$
2) Вычислить обратную матрицу $A^{-1} = C$
3) $T = Tr(C)$, $R = ∑c_{ij}$
Тогда центральность по информации:

$$C_I(i) = \frac{1}{c_{ii} + (T-2R)/g}$$

Нормированная центральность по информации:

$$C_I'(i) = \frac{C_I(i)}{∑C_I(j)}$$

### Групповые характеристики сетей

#### Централизация по степени
$$
C_D = \frac{\sum_{i=1}^g [C_D(n^*) - C_D(n_i)]}{(g-1)(g-2)}
$$
где $C_D(n^*)$ - максимальная степень центральности, $g$ - число вершин.

#### Централизация по близости
$$
C_C = \frac{\sum_{i=1}^g [C'_C(n^*) - C'_C(n_i)]}{[(g-2)(g-1)]/(2g-3)}
$$

#### Централизация по посредничеству
$$
C_B = \frac{\sum_{i=1}^g [C'_B(n^*) - C'_B(n_i)]}{g-1}
$$


### Меры престижа для ориентированных сетей

#### Престиж по степени
$$P_D(i) = \frac{deg_i(i)}{g-1}$$

где $deg_i(i)$ - полустепень захода.

#### Престиж по близости
$$
P_P(i) = \frac{|I_i|}{g-1} \cdot \frac{|I_i|}{\sum d_{ij}}
$$
где $|I_i|$ — количество вершин, достижимых из $i$,  

#### Престиж по рангу
Решается система уравнений:
$$
\left(\frac{1}{a}E - X\right)^TP_R = x
$$
где:  
- $X$ — социоматрица
- $a = \frac{1}{\lambda}$, где $\lambda$ — максимальное собственное значение матрицы $X$  
- $x$ — вектор сумм столбцов  
- $E$ — единичная матрица  

Нормированный престиж:
$$
P_R^{'}(n_i) = \frac{P_R(n_i)}{m}$$

где $ m = \sum^{g-1}_{k=1} a^kA^k_{g-1}$


### Алгоритмы визуализации сетей

#### Фрюхтерман-Рейнгольд
Основан на физической аналогии:
- Вершины отталкиваются друг от друга
- Рёбра действуют как пружины
Минимизирует функцию энергии:

$$E = \sum_{i \neq j} k_r(d_{ij}) + \sum_{(i,j) \in E} k_a(d_{ij})$$

где $k_r$ - сила отталкивания, $k_a$ - сила притяжения.

#### Камада-Каваи
Минимизирует разницу между геодезическими и евклидовыми расстояниями:

$$E = \sum_{i \neq j} \frac{1}{2} k_{ij}(||p_i - p_j|| - l_{ij})^2$$

где $l_{ij}$ - желаемое расстояние, $k_{ij}$ - вес пружины.

#### 1. Извлечение из florentine_business.csv социоматрицы



```r
# Переменная хранит адрес файла florentine_business.csv
table <- read.csv('./florentine_business.csv', header = FALSE, skip = 2)

graph <- igraph::graph_from_adjacency_matrix(as.matrix(table), mode ="undirected")
network <- igraph::as_adjacency_matrix(graph)

names <- colnames(read.csv("./florentine_business.csv", skip = 1))

rownames(network) <- names
colnames(network) <- names
network

plot(graph, layout = layout_with_fr(graph), displaylabels = TRUE)
```

#### 2. Поиск точек сочленения и мостов


```r
articulation_points <- articulation_points(graph)
cat("Точки сочленения:", V(graph)$name[articulation_points], "\n")
```

```r
bridges <- bridges(graph)
vertex_colors <- rep("lightblue", vcount(graph))
vertex_colors[articulation_points] <- "red"


colors <- rep("gray80", ecount(graph))
colors[bridges] <- "darkblue"

widths <- rep(1, ecount(graph))
widths[bridges] <- 3

plot(graph, 
     layout = layout_with_fr(graph),
     vertex.color = vertex_colors,
     vertex.label.color = "black",
     vertex.size = 15,
     edge.color = colors,
     edge.width = widths,
     main = "Красные вершины - это точки сочленения, а темносиние ребра - мосты")
```

#### 3.  Значение централизации для каждой меры

```r
# Удаляем изолированные вершины (используем igraph::degree)
non_isolated_vertices <- which(igraph::degree(graph) > 0)
g <- igraph::induced_subgraph(graph, non_isolated_vertices)
n <- igraph::vcount(g)

# Меры центральности с указанием пакета
deg <- igraph::degree(g)
clo <- igraph::closeness(g, mode = "all", normalized = TRUE)
bet <- igraph::betweenness(g, directed = FALSE)

# Централизация сети
C_D <- sum(max(deg) - deg) / ((n - 1) * (n - 2))
C_C <- sum(max(clo) - clo) * (2*n - 3) / ((n - 1) * (n - 2))
C_B <- 2 * sum(max(bet) - bet) / ((n - 1)^2 * (n - 2))

# Вывод результатов
cat("Централизация по степени: ", round(C_D, 4), "\n")
cat("Централизация по близости: ", round(C_C, 4), "\n")
cat("Централизация по посредничеству:", round(C_B, 4), "\n")
```

##### Вывод
1.Централизация по степени — 0.2778 (Средняя). Это означает, что в сети есть несколько узлов с относительно большим числом прямых связей, но сеть не имеет ярко выраженного центра (не звёздная структура). Распределение связей между узлами умеренное, что указывает на смешанную структуру сети без единственного доминирующего узла.

2.Централизация по близости — 0.03535 (Низкая). Это говорит нам, что большинство узлов имеют примерно одинаковую "доступность" к остальным: средняя длина кратчайших путей от одной вершины до всех остальных слабо вариативна. Это говорит о хорошей связности, отсутствии явно выделенных «центральных» узлов, которые значительно быстрее достигают другие.

3.Централизация по посредничеству — 0.4422 (Относительно высокая). Значение выше 0.4 говорит о наличии узлов, которые лежат на кратчайших путях между другими узлами — своеобразных «посредников» или «мостов». Удаление таких ключевых узлов может значительно повлиять на связность и поток информации по сети.

### Работа с сетью аффелированости ceo.csv

#### 6. Извлечение из ceo.csv социоматрицы

```r
data <- read.csv("./ceo.csv", header = FALSE, skip = 9, sep = ",")

graph1 <- igraph::graph_from_biadjacency_matrix(as.matrix(data))
net1 <- igraph::as_biadjacency_matrix(graph1)

net1
```

#### 7. Отображение сети в виде биграфа

```r
plot(graph1, layout=layout.bipartite)
```

#### 8. Отображение сети в виде гиперграфа



```r
types <- V(graph1)$type

# Создаем гиперрёбра за один проход
hyperedges <- sapply(which(types), function(club) {
  ceos <- neighbors(graph1, club)[!types[neighbors(graph1, club)]]
  if (length(ceos) > 0) as.character(ceos) else NULL
}, simplify = FALSE)

# Убираем NULL и задаём имена
hyperedges <- Filter(Negate(is.null), hyperedges)
names(hyperedges) <- paste("Club", which(types)[sapply(hyperedges, function(x) !is.null(x))])

# Создаём и визуализируем гиперграф
plot(as.hypergraph(hyperedges), main = "Гиперграф сети")
```

#### 9. Представление сети аффилированности в виде двух одномодальных сетей



```r
# Создание одномодальных сетей
A <- net1 %*% t(net1); diag(A) <- 0
B <- t(net1) %*% net1; diag(B) <- 0

g1 <- graph_from_adjacency_matrix(A, "undirected", weighted = TRUE)
g2 <- graph_from_adjacency_matrix(B, "undirected", weighted = TRUE)

# Вывод матриц
print("Матрица акторов:"); print(A)
print("Матрица событий:"); print(B)

# Визуализация
par(mfrow = c(1, 2))
plot(g1, main = "Акторы", vertex.size = 8, edge.width = E(g1)$weight/2)
plot(g2, main = "События", vertex.size = 8, edge.width = E(g2)$weight/2)
par(mfrow = c(1, 1))
```

##### Вывод
Анализ показывает, что связи между CEO и между клубами распределены неравномерно. Выделяется тесная группа (ядро), где одни и те же руководители входят в одни и те же клубы. Это видно по "тяжёлым" связям в матрицах и густым пучкам на графах. Такая картина типична для сплочённых элитных кругов. В то же время, есть и слабые, редкие связи на периферии сети. Итог: структура аффилиационной сети явно делится на влиятельное ядро и менее активную периферию, что характерно для мира корпоративного управления.

#### 14. Определение логарифма отношения шансов.


```r
# Компактный расчет логарифма отношения шансов

n <- ncol(net1)
logOR <- matrix(0, n, n)

for (i in 1:(n-1)) {
    for (j in (i+1):n) {
        a <- sum(net1[,i] & net1[,j]) + 0.5
        b <- sum(net1[,i] & !net1[,j]) + 0.5
        c <- sum(!net1[,i] & net1[,j]) + 0.5
        d <- sum(!net1[,i] & !net1[,j]) + 0.5
        
        logOR[i,j] <- logOR[j,i] <- log((a*d)/(b*c))
    }
}

# Результаты
print(round(logOR, 2))

# Интерпретация
cat("\nИнтерпретация logOR:\n")
cat("> 0: притяжение между клубами (совместное участие чаще случайного)\n")
cat("< 0: отталкивание между клубами (совместное участие реже случайного)\n")
cat("= 0: независимость (совместное участие как ожидалось)\n")
```

##### Вывод
 При подсчете шансов была добавлена поправка Холдейна-Энскомба, чтобы избавиться от деления на ноль.
 Анализ логарифма отношения шансов показывает, что связи между клубами носят неоднородный характер: Наличие положительных значений  указывает на сильную склонность к совместному членству.
Отрицательные значения свидетельствуют о структурной изоляции: участники одного клуба редко входят в другой, что может отражать различия в отраслевой, функциональной или социальной ориентации.
В целом, сеть демонстрирует смешанную структуру: существуют как тесно связанные так и разобщённые подсистемы.

Проведённый анализ сети, объединяющей 26 руководителей и 15 клубов, выявил устойчивую структуру. Преобразование в две одномодальные сети показало наличие плотного ядра взаимосвязанных элементов. Все три метода кластеризации согласованно выделили 7 сообществ с модулярностью Q ≈ 0,364, что подтверждает объективность структуры. Среднее участие CEO составляет 4–5 клубов, а анализ логарифма отношений шансов указал на наличие как тесно связанных, так и разделённых пар клубов, что отражает их функциональную специализацию.
