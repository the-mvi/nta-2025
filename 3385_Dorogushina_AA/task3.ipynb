## Основные теоретические положения

### Основные определения

#### Актор
Актор (actor) - базовый элемент сети, представляющий отдельную единицу, (не обязательно) участвующую во взаимодействиях.

#### Реляционная связь
Реляционная связь (relational tie) - отношение/взаимодействие/ассоциация между двумя акторами в сети.

#### Граф
Граф (graph) - модель социальной сети с ненаправленным дихотомическим отношением $G = G(N,L)$, где:
- $N = {n_1,...,n_g}$ - набор акторов;
- $L = {l_1,...,l_L}$, $l_k = (n_i,n_j)$ - набор ребер.

#### Социоматрица
Социоматрица (sociomatrix) - матрица $X$ размерности $g \times g$, описывающая смежные узлы в сети ($g$ - количество акторов в сети). $X = x_{ij}, 1 \le i,j \le g$, т.ч.
- на диагонали $x_{ii} = -$;
- в остальных  $x_{ij} = 0$ или  $x_{ij} = 1$.

#### Размер сети
Резмер сети (network size) - количество акторов в сети.

#### Плотность сети
Плотность сети (network density) - доля имеющихся связей по отношению к максимально возможному количеству связей:

$$den(G) = \frac{2L}{g(g-1)} = \frac{L}{\binom{g}{2}}$$

#### Компонента связности
Компонента связности (component) - подгруппа, в которой все акторы связаны друг с другом.

#### Диаметр сети
Диаметр сети (network diameter) - максимальное расстояние между 2 узлами в сети:
$$diam(G) = \max d(n_i,n_j)$$

- $d_{ij}$ - длина кратчайшего пути между вершинами $i$ и $j$.

#### Двудольные сети
Двудольный граф (bipartite graph) - сеть $G = (V, E)$, где $V = V_1 \cup V_2$, $V_1 \cap V_2 = \emptyset$, и все рёбра $e \in E$ соединяют вершины из разных множеств:
$$
E \subseteq \{(u,v) | u \in V_1, v \in V_2\}
$$


#### Гиперграф
Обобщение графа, в котором каждым ребером могут соединяться не два узла, а подмножества

$$
H = H(A, B)$$ где
- $A$ - набор вершин
- $B$ - набор гипер ребер

#### Матрицы
- Матрица совместного участия акторов $X_N = A A^T$
- Матрица перекрытия событий $X_M = A^T A$

#### K-ядра
$k$-ядро ($k$-core) - максимальный подграф, в котором каждый узел смежен не менее чем с k узлами в подграфе:
$$\deg(v) \geq k, ∀ v \in V$$


#### Отношение шансов
Пусть $kl$ - количество акторов, которые участвуют в $k$ и $l$ событиях.
Аналогично $k$ и $l$.
$0$ - количество акторов, не участвующих в событиях

Тогда отношение шансов и его логарифм:
- $O_kl = \frac{kl \cdot 0}{k \cdot l}$
- $ln(O_kl)$

### Групповые характеристики сетей
$C_A(n^*) = max_iC_A(n_i)$
#### Централизация по степени
$$
C_D = \frac{\sum_{i=1}^g [C_D(n^*) - C_D(n_i)]}{(g-1)(g-2)}
$$

#### Централизация по близости
$$
C_C = \frac{\sum_{i=1}^g [C'_C(n^*) - C'_C(n_i)]}{[(g-2)(g-1)]/(2g-3)}
$$

#### Централизация по посредничеству
$$
C_B = \frac{2 \sum_{i=1}^g [C'_B(n^*) - C'_B(n_i)]}{(g-2)(g-1)^2}
$$

### Диады и цензус диад

#### Диада
Диада (dyad) — неупорядоченная пара акторов и направленных ребер, которые существуют между ними
$$ D_{ij} = (x_{ij}, x_{ji})$$

- симметричная (mutual): дуги $  i \to j $ и $ j \to i $ присутствуют одновременно - (1,1)

$$ M = 1/2 * tr(X^2)$$

- ассиметричная (asymmetric): присутствует ровно одна из двух возможных дуг - (1,0) или (0,1)

$$ A = tr(XX^T) - 2M$$

- нулевая (null): обе дуги отсутствуют - (0,0)

$$ N = C^2_g - A - M $$

#### Цензус диад
Цензус диад (dyad census) — тройка объектов $ (M, A, N) $

- $M+A+N = g(g-1)$

### Индекс взаимности Каца–Пауэлла

Характеристика измерения тенденции акторов в группе
$$
\mu_{\text{кр}} = \frac{2(g - 1)M - gd^2}{gd(g - d -1)},
$$

где $d ≤ g - 1$ - число выборов, который может сделать каждый актор

- фиксированная модель: предполагается, что каждый узел имеет одинаковую ожидаемую степень $ d $
- свободная модель: учитывает реальное распределение входящих и исходящих степеней

#### Вероятность взаимной связи

В рамках модели Каца–Пауэлла оценивается вероятность того, что диада $ ij $ сопровождается диадой $ ji $. Вычисляется отдельно для:
- фиксированного выбора
- свободного выбора

### Индекс взаимности Бондопаддхая

Индекс, основанный на минимальном и максимальном значении количества взаимных диад:
$$
M_B = \frac{M - M_{\min}}{M_{\max} - M_{\min}},
$$
где:
- $ M $ — наблюдаемое число взаимных диад
- $ M_{\min} ,  M_{\max} $ — минимально и максимально возможное число взаимных диад при данном распределении степеней

###1. Моделирование случайного орграфа

```r
# сначала добавление нужных библиотек
if (!require(statnet)) install.packages("statnet", repos = "https://cloud.r-project.org/")
if (!require(igraph)) install.packages("igraph", repos = "https://cloud.r-project.org/")
```

```r
library(statnet)
library(igraph)
```

```r
g <- 94      # количество акторов
p <- 0.58    # вероятность появления дуги

# создание случайного орграфа по модели Эрдеша-Реньи
G_d <- sample_gnp(n = g, p = p, directed = TRUE)

# матрица смежности
cat("\nФрагмент матрицы смежности (первые 5 строк и столбцов):\n")
A <- as_adjacency_matrix(G_d)
print(A[1:5, 1:5])
```

###2. Отображение полученного орграфа

```r
cat("Случайный орграф G_d = G(", g, ", ", p, ")\n", sep = "")
cat("Количество вершин (акторов):", vcount(G_d), "\n")
cat("Количество дуг:", ecount(G_d), "\n")
cat("Плотность графа:", edge_density(G_d), "\n")

plot(G_d,
  vertex.size = 8,
  vertex.label.cex = 0.7,
  edge.arrow.size = 0.2,
  edge.width = 0.2,
  main = paste("Случайный орграф G(", g, ", ", p, ")", sep = ""))
```

###3. "Пятичисловая сводка" графа $G_d$

```r
size_g <- vcount(G_d)
cat("1. Размер графа:", size_g, "\n\n")

cat("2. Плотность графа:\n")
max_edges <- size_g * (size_g - 1)
actual_edges <- ecount(G_d)
den_g_manual <- actual_edges / max_edges
cat("   Вручную по формуле:", den_g_manual, "\n")
cat("   По функции:", edge_density(G_d), "\n")
cat("   Ожидаемая плотность (параметр p):", p, "\n\n")

cat("3. Компоненты связности:\n")
cat("   Число сильно связных компонент:", count_components(G_d, mode = "strong"), "\n")
cat("   Число слабо связных компонент:", count_components(G_d, mode = "weak"), "\n\n")

cat("4. Диаметр графа:\n")
gd <- diameter(G_d)
cat("   Диаметр:", gd, "\n")
avg_path_length <- mean_distance(G_d)
cat("   Средняя длина пути:", avg_path_length, "\n\n")

cat("5. Коэффициент кластеризации:\n")
trans_global <- transitivity(G_d, type = "global")
trans_average <- transitivity(G_d, type = "average")
cat("   Глобальный коэффициент кластеризации:", trans_global, "\n")
cat("   Средний локальный коэффициент кластеризации:", trans_average, "\n")
```

Нулевых меньше, чем симметричных, т.к. плотность чуть больше 1/2

###5. Вычислить оценку индекса взаимности Каца—Пауэла при фиксированом выборе акторов 𝑑 и при свободном выборе

```r
d = (g-1) * p # исходящая степень
fix <- (2 * (g - 1) * d_c$mut - g * d * d) / (g * d * (g - d - 1))
cat("Оценка индекса взаимности КП с фикс d =", d, ":\n", fix, "\n\n")

degree_dist <- degree(G_d)
degree_list_out <- degree(G_d, mode = "out")
L <- sum(degree_list_out)
L2 <- sum(degree_list_out^2)
no_fix <- (2 * (g - 1) * (g - 1) * d_c$mut - L * L + L2) / (L * (g - 1) * (g - 1) - L * L + L2)
cat("Оценка индекса взаимности КП без фикса :\n", no_fix)
```

Выводы:

В сети взаимных связей чуть больше, чем ожидалось в случайной модели с теми же параметрами с теми же параметрами, но далеко не максимальная взаимность
- R = 0 -> взаимность как в случайной сети,
- R > 0 -> избыток взаимности,
- R < 0 -> дефицит

###6. Вычислить вероятность того, что связь будет взаимной для каждого случая

```r
mut_fix <- (d / (g - 1)) * ((d / (g - 1)) + fix * (g - d - 1) / (g - 1))
cat("Вероятность взаимной связи с фикс d:\n", mut_fix, "\n\n")
mut_no_fix <- (L / (g * (g - 1))) * ((L / (g * (g - 1)))  + no_fix * (1 - (L / (g * (g - 1)))))
cat("Вероятность взаимной связи без фикса:\n", mut_no_fix)
```

###7. Вычислить индекс взаимности Бондопаддхая

```r
max_g <- -Inf
M_min <- -Inf

for (t in 1:g) {
    f_t <- sum(degree_list_out[1:t]) - t * (g - 1) - (t * (t - 1)) / 2
    if (t == g)
        sum_min_rest <- 0
    else
        sum_min_rest <- sum(pmin(t, degree_list_out[(t + 1):g]))
    g_t <- sum(degree_list_out[1:t]) - t * (t - 1) - sum_min_rest
    if (g_t > max_g) max_g <- g_t
    if (f_t > M_min) M_min <- f_t
}

M_min <- pmax(0, M_min)
M_max <- (sum(degree_list_out) - max_g) / 2

cat("M_max: ", M_max, "\n")
cat("M_min: ", M_min, "\n")
M_b <- (d_c$mut - M_min) / (M_max - M_min)
cat("Индекс взаимности Бондопаддхая: ", M_b, "\n")
```

Выводы:

Взаимность чуть больше средней
- M_b ~ 0 -> минимальная взаимность
- M_b ~ 1 -> максимальная взаимность
- M_b ~ 0.5 -> средняя взаимность

###8. Проверить нулевую гипотезу. Определить максимальный уровень значимости при котором нулевая гипотеза не отвергается ($1/2$, $\alpha$=0,10)

```r
L <- ecount(G_d)
n <- g * (g - 1)

test_binom <- function(L, n, p0, alpha = 0.10) {
  bt <- binom.test(L, n, p = p0, alternative = "two.sided")
  cat("p0 =", p0, "\n")
  cat("p-value =", bt$p.value, "\n")
  cat("Нулевая гипотеза", ifelse(bt$p.value < alpha, "отвергается", "не отвергается"),
      "на уровне α =", alpha, "\n")
  cat("p-value = max α = ", bt$p.value)
}
```

```r
test_binom(L, n, p0 = 0.5)
```

###9. Проверить нулевую гипотезу. Определить максимальный уровень значимости при котором нулевая гипотеза не отвергается ($P_0 = 0,73$)

```r
test_binom(L, n, p0 = 0.73)
```

###10. Построить доверительный интервал для оценки параметра 𝑝 с доверительной надёжностью 1 − $\alpha$

```r
p_hat <- L / n
se <- sqrt(p_hat * (1 - p_hat) / n)
z <- qnorm(1 - alpha / 2)
ci_low <- max(0, p_hat - z * se)
ci_up  <- min(1, p_hat + z * se)

cat("Оценка p =", p_hat, "\n")
cat(sprintf("%.0f%% доверительный интервал для p: [%.4f, %.4f]\n", (1 - alpha) * 100, ci_low, ci_up))
cat("p = 0.58", ifelse(ci_low <= 0.58 && 0.58 <= ci_up, "попадает", "не попадает"), "в доверительный интервал\n")
```

Выводы:

90% доверительный интервал [0.5744, 0.5917] достаточно узкий

Реальное значение p = 0.76 лежит внутри этого интервала
