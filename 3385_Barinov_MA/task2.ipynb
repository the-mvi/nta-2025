## Основные теоретические положения

### Основные определения

#### Социоматрица
Социоматрица $A$ - это квадратная матрица размера $g \times g$, $g$ - количество акторов в сети. А каждый элемент  принимает значение:
- $1$ - существует связь от актора $i$ к актору $j$
- $0$ - если связь отсутствует


#### Список рёбер
Альтернативное представление сети в виде множества упорядоченных пар ребер:
$E = \{(i,j) | a_{ij} = 1\}$

#### Плотность сети
Плотность ненаправленной сети G вычисляется как:

$$den(G) = \frac{2L}{g(g-1)}$$

где $L$ - количество имеющихся связей, $g$ - количество акторов.

#### Компоненты связности
Компонента связности - максимальное подмножество вершин, где существует путь между любой парой вершин.

#### Точка сочленения
Такая вершина при удалении которой из графа он распадается на несколько компонент связности.

#### Мост
Такое ребро при удалении которого из графа он распадается на несколько компонент связности.

#### Диаметр сети
Диаметр $diam$ - максимальное геодезическое расстояние между любой парой вершин в сети G:

$$diam(G) = \max_{i,j} d_{ij}$$

где $d_{ij}$ - длина кратчайшего пути между вершинами $i$ и $j$.

#### Коэффициент кластеризации
Локальный коэффициент кластеризации $Cl_i$ для вершины $i$ графа G:

$$Cl_i(G) = \frac{2N_i}{k_i(k_i-1)}$$

где $N_i$ - количество связей между соседями узла $i$, $k_i$ - степень узла.

Глобальный коэффициент кластеризации:

$$Cl(G) = \frac{3 * T_t(G)}{T_c(G)}$$

где $T_t$ - число треугольников, а $T_c$ - число открытых и закрытых треугольников.

#### Двудольные сети

Сеть $G = (V, E)$, где $V = V_1 \cup V_2$, $V_1 \cap V_2 = \emptyset$, и все рёбра $e \in E$ соединяют вершины из разных множеств:
$$
E \subseteq \{(u,v) | u \in V_1, v \in V_2\}
$$


#### Гиперграф

Это обобщение графа, в котором каждым ребером могут соединяться не два узла, а произвольное подмножество вершин графа

$$
H = H(A, B)$$ где

$A$ - набор вершин

$B$ - набор гипер ребер


#### Проекции двудольной сети

- *Проекция на акторы* ($P_1$): $A_1 = B B^T - D_1$
- *Проекция на события* ($P_2$): $A_2 = B^T B - D_2$

где $D_1$, $D_2$ - диагональные матрицы ($B B^T$ и $B^T B$ соответственно).


#### K-ядра

$k$-ядро - максимальный подграф, в котором каждый узел смежен не менее чем с k узлами в подграфе:

$$\deg(v) \geq k, ∀ v \in V$$

#### Клика

Клика это максимальный подграф.

#### Отношение шансов

Для таблицы сопряженности:

| | Участвует в $j$ | Не участвует в $j$ |
|-|-----------------|-------------------|
| Участвует в $i$ | $a$ | $b$ |
| Не участвует в $i$ | $c$ | $d$ |

$$
OR = \frac{a \cdot d}{b \cdot c}, \quad LOR = \ln(OR)
$$


### Меры центральности

#### Центральность по степени
Для вершины $i$ в неориентированной сети:

$$C_D(i) = deg(i)$$

Нормированная центральность по степени:

$$C_D'(i) = \frac{deg(i)}{g - 1}$$

#### Центральность по близости
$$C_C(i) = \frac{1}{\sum_{j \neq i} d_{ij}}$$

где $d_{ij}$ - расстояние между вершинами.

Нормированная центральность по близости:

$$C_C'(i) = \frac{g-1}{\sum_{j \neq i} d_{ij}}$$

#### Центральность по посредничеству
$$C_B(i) = \sum_{j < k} \frac{\sigma_{jk}(i)}{\sigma_{jk}}$$

где $\sigma_{jk}$ - количество кратчайших путей между $j$ и $k$, $\sigma_{jk}(i)$ - количество таких путей через $i$.

Нормированная центральность по посредничеству:

$$C_B'(i) =\frac{2 C_B(i)}{(g-1)(g-2)}$$

#### Центральность по информации

Алгоритм нахождения:
1) Вычислить матрицу псевдосмежности $A = 1 + K$, где $K$ - матрица Кирхгофа $g \times g$
2) Вычислить обратную матрицу $A^{-1} = C$
3) $T = Tr(C)$, $R = ∑c_{ij}$
Тогда центральность по информации:

$$C_I(i) = \frac{1}{c_{ii} + (T-2R)/g}$$
$Tr(C)$ - след матрицы(сумма элементов на главной диагонали)

Нормированная центральность по информации:

$$C_I'(i) = \frac{C_I(i)}{∑C_I(j)}$$


### Меры престижа для ориентированных сетей

#### Престиж по степени
$$P_D(i) = \frac{deg_{in}(i)}{g-1}$$


#### Престиж по близости
$$
P_P(i) = \frac{|I_i|}{g-1} \cdot \frac{|I_i|}{\sum d_{ij}}
$$
где $|I_i|$ — количество вершин, достижимых из $i$,  

#### Престиж по рангу
Получается через решение системы уравнений:
$$
\left(\frac{1}{a}E - X\right)^TP_R = x
$$
где:  
- $X$ — социоматрица
- $a = \frac{1}{\lambda}$, где $\lambda$ — максимальное собственное значение матрицы $X$  
- $x$ — вектор сумм столбцов  
- $E$ — единичная матрица  

Нормированный престиж:
$$
P_R^{'}(n_i) = \frac{P_R(n_i)}{m}$$

где $ m = \sum^{g-1}_{k=1} a^kA^k_{g-1}$


### Меры централизации
Меры централизации выявляют ключевые вершины в сети и оценивают концентрацию влияния в графе.

#### Централизация по степени
Локальное влияние
$$
C_D = \frac{\sum_{i=1}^g [C_D(n^*) - C_D(n_i)]}{(g-1)(g-2)}
$$
где $C_D(n^*)$ - максимальная степень центральности, $g$ - число вершин.

#### Централизация по близости
Скорость распространения
$$
C_C = \frac{\sum_{i=1}^g [C'_C(n^*) - C'_C(n_i)]}{[(g-2)(g-1)]/(2g-3)}
$$

#### Централизация по посредничеству
$$
C_B = \frac{\sum_{i=1}^g [C'_B(n^*) - C'_B(n_i)]}{g-1}
$$

### Алгоритмы визуализации сетей

#### Фрюхтерман-Рейнгольд
Работает по принципу пружинок, если ребро есть то вершины стараются притянуться.
Минимизирует функцию энергии:

$$E = \sum_{i \neq j} k_r(d_{ij}) + \sum_{(i,j) \in E} k_a(d_{ij})$$

где $k_r$ - сила отталкивания, $k_a$ - сила притяжения.

#### Камада-Каваи

Цель алогритма разместить вершины так, чтобы расстояние на "рисунке" соотвествовало графовым расстояниям.

Минимизирует разницу между геодезическими и евклидовыми расстояниями:

$$E = \sum_{i \neq j} \frac{1}{2} k_{ij}(||p_i - p_j|| - l_{ij})^2$$

где $l_{ij}$ - желаемое расстояние, $k_{ij}$ - вес пружины.

### Задание 3.

Вычислить значение централизации для каждой меры. Сделать выводы

```r
non_isolated_vertices <- which(degree(graph) > 0)
g <- induced_subgraph(graph, non_isolated_vertices)

deg <- degree(g)

# Близость (Closeness Centrality)
closeness_cent <- closeness(g, mode = "all")

# Посредничество (Betweenness Centrality)
betweenness_cent <- betweenness(g, directed = FALSE)

n_vertices <- vcount(g)

C_D <- sum(max(deg) - deg) / ((n_vertices - 1) * (n_vertices - 2))
C_C <- (sum(max(closeness_cent) - closeness_cent) * (2*n_vertices - 3)) / ((n_vertices - 1) * (n_vertices - 2))
C_B <- sum(max(betweenness_cent) - betweenness_cent) / ((n_vertices - 1) * (n_vertices - 2) * (n_vertices - 3) / 2)
cat("C_C (степень) =", round(C_D, 4), "\n")
cat("C_C (близость) =", round(C_C, 4), "\n")
cat("C_B (посредничество) =", round(C_B, 4), "\n")
```

#### Вывод

1. Централизация по степени равна 0.2778, что чуть ниже среднего уровня. Такая величина указывает на наличие нескольких узлов с повышенным числом прямых связей, при этом сеть не демонстрирует структуры с единственным доминантным центром, характерной для звёздной топологии. Распределение связей характеризуется умеренной неоднородностью, что свидетельствует о смешанной архитектуре сети без явного преобладающего элемента.

2. Показатель централизации по близости составляет 0.0354, что является очень низким значением. Это отражает относительно равномерную доступность узлов друг к другу, проявляющуюся в небольших различиях средней длины кратчайших путей между узлами. Это говорит о высокой связанности сети и отсутствии узлов с выраженным преимуществом в скорости достижения других.

### Задание 5.

С помощью библиотеки igraph исследовать алгоритмы обнаружения сообществ в зависимости от номера варианта:
- (1) Edge-betweenness, Leading eigenvector, Fast-greedy;

```r
communities_eb <- edge.betweenness.community(graph)
communities_le <- leading.eigenvector.community(graph)
communities_fg <- fastgreedy.community(graph)

eb_str <- paste("Edge-Betweenness: сообществ =", length(communities_eb), ", Q =", round(modularity(communities_eb), 4))
le_str <- paste("Leading Eigenvector: сообществ =", length(communities_le), ", Q =", round(modularity(communities_le), 4))
fg_str <- paste("Fast-Greedy: сообществ =", length(communities_fg), ", Q =", round(modularity(communities_fg), 4))

cat(eb_str, "\n", le_str, "\n", fg_str, "\n\n")

plot(communities_eb, graph, main = eb_str, vertex.size = 8, vertex.label.cex = 0.6)
plot(communities_le, graph, main = le_str, vertex.size = 8, vertex.label.cex = 0.6)
plot(communities_fg, graph, main = fg_str, vertex.size = 8, vertex.label.cex = 0.6)

par(mfrow = c(1, 1))
```

#### Вывод

Алгоритмы обнаружения сообществ.
Основная задача таких алгоритмов найти в сети естественные групировки узлов, которые сильнее связаны между собой чем с остальной сетью.

`Edge-Betweenness` — это алгоритм разделения графа, который работает путём последовательного удаления рёбер с наибольшим значением посредничества (betweenness centrality). Идея заключается в том, что рёбра, которые часто лежат на кратчайших путях между вершинами из разных сообществ, играют роль "мостов" между группами. Удаляя такие мосты, граф разбивается на всё более мелкие компоненты, которые и являются сообществами. Алгоритм работает медленнее других, но даёт хороший результат для малых и средних сетей.

`Leading Eigenvector` — это спектральный метод, основанный на анализе собственных векторов матрицы модулярности. Алгоритм ищет такое разбиение вершин на сообщества, которое максимизирует модулярность (меру качества разбиения). Он использует математический аппарат линейной алгебры для нахождения оптимального разделения. Этот метод даёт часто оптимальные или близкие к оптимальным результаты, но работает только для связных графов.

`Fast-Greedy` — это жадный алгоритм, который начинает с того, что каждая вершина — отдельное сообщество, а затем последовательно объединяет две сообщества, если это увеличивает модулярность. На каждом шаге выбирается пара сообществ, которая даёт максимальный прирост модулярности. Алгоритм работает очень быстро и часто находит хорошие результаты, хотя теоретически может застревать в локальных оптимумах.

### Задание 14.

Для каждых пар событий определить (логарифм) отношение шансов. Сделать выводы

```r
n_ceo <- nrow(network1)
n_clubs <- ncol(network1)


logOR <- matrix(0, nrow = n_clubs, ncol = n_clubs)


for (i in 1:(n_clubs - 1)) {
  for (j in (i + 1):n_clubs) {
    a <- sum(network1[, i] & network1[, j]) + 0.5   # в обоих клубах
    b <- sum(network1[, i] & !network1[, j]) + 0.5  # только в i
    c <- sum(network1[, j] & !network1[, i]) + 0.5  # только в j
    d <- sum(!network1[, i] & !network1[, j]) + 0.5 # ни в одном

    or <- (a * d) / (b * c)
    logOR[i, j] <- logOR[j, i] <- round(log(or), 2)
  }
}
print(logOR)
```

#### Вывод

Матрица показывает неоднородные ассоциации между событиями. Сильные положительные связи (LOR > 1.5) наблюдаются между событиями 1-4, 4-10, 6-15, 11-15, что указывает на кластеры взаимосвязанных событий с общей аффилиацией. Сильные отрицательные связи (LOR < -1.5) между 14-4 и 14-15 показывают разделённые события, которые не пересекаются по акторам. Большинство пар имеют LOR около нуля, свидетельствуя о независимости. Такая структура характерна для сети с несколькими плотными ядрами и изолированными группами.
