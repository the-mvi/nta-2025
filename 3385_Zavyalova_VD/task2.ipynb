## Основные теоретические положения


## 1. Матричные представления

### Социоматрица (матрица смежности)
Социоматрица $A$ — квадратная матрица размера $g \times g$, где $g$ — число акторов (вершин) сети.

Элемент $a_{ij}$:
- $1$ — если существует связь от актора $i$ к актору $j$
- $0$ — если связи нет

Для **неориентированных** сетей матрица симметрична:
$$ a_{ij} = a_{ji} $$

### Список рёбер
Альтернативное представление сети как множества установленных связей:
$$ E = \{ (i, j) \mid a_{ij} = 1 \} $$



## 2. Общие сетевые меры

### Плотность сети
Для неориентированной сети $G$:
$$ \text{den}(G) = \frac{2L}{g(g - 1)} $$

где:
- $L$ — количество рёбер
- $g$ — количество вершин

### Компоненты связности
**Компонента связности** — максимальный подграф, в котором существует путь между любой парой вершин.

### Диаметр сети
$$ \text{diam}(G) = \max_{i, j} \ d_{ij} $$

где $d_{ij}$ — длина кратчайшего пути между вершинами $i$ и $j$.



## 3. Меры локальной структуры

### Коэффициент кластеризации

#### Локальный коэффициент
$$ Cl_i(G) = \frac{2N_i}{k_i(k_i - 1)} $$

где:
- $k_i$ — степень вершины $i$
- $N_i$ — количество треугольников, проходящих через вершину $i$

#### Глобальный коэффициент
$$ Cl(G) = \frac{1}{n} \sum_{i=1}^n Cl_i $$

### $k$-ядра
$k$**-ядро** — максимальный подграф, в котором степень каждой вершины не меньше $k$:
$$ \deg(v) \ge k \quad \forall v \in V $$



## 4. Сети аффилиации (двудольные сети)

### Определение
Граф $G = (V, E)$ является **двудольным**, если:
$$ V = V_1 \cup V_2, \quad V_1 \cap V_2 = \varnothing $$
и все рёбра соединяют вершины из разных долей:
$$ E \subseteq \{ (u, v) \mid u \in V_1,\ v \in V_2 \} $$

> **Пример:** Акторы ($V_1$) и события ($V_2$), в которых они участвуют.

### Матрица инцидентности
Обозначается $B$ (размер $|V_1| \times |V_2|$). Элемент $b_{ij} = 1$, если актор $i$ участвует в событии $j$.

### Проекции двудольной сети

#### Проекция на акторов ($P_1$)
$$ P_1 = B B^\top - D_1 $$
Два актора связаны, если участвуют хотя бы в одном общем событии.


#### Проекция на события ($P_2$)
$$ P_2 = B^\top B - D_2 $$
Два события связаны, если в них участвует хотя бы один общий актор.

где $D_1$, $D_2$ — диагональные матрицы степеней (убирают петли).


### Гиперграф
Обобщение графа, где рёбра (гиперрёбра) могут соединять любое подмножество вершин:
$$ H = (A, B) $$
где:

- $A$ — множество вершин
- $B$ — множество гиперрёбер



## 5. Статистические меры

### Отношение шансов (Odds Ratio)


#### Таблица сопряжённости

| | Участвует в $j$ | Не участвует в $j$ |
|-|-----------------|-------------------|
| **Участвует в $i$** | $a$ | $b$ |
| **Не участвует в $i$** | $c$ | $d$ |


#### Формулы
$$ OR = \frac{a \cdot d}{b \cdot c} $$
$$ LOR = \ln(OR) $$

> **Интерпретация:** $OR > 1$ означает положительную связь между участием в событиях $i$ и $j$.


## Меры центральности

### 1. Центральность по степени

Для вершины $(i$) в неориентированной сети:

$$
C_D(i) = \deg(i)
$$

Нормированная центральность по степени:

$$
C_D'(i) = \frac{\deg(i)}{g - 1}
$$



### 2. Центральность по близости

$$
C_C(i) = \frac{1}{\sum_{j \neq i} d_{ij}}
$$

где $(d_{ij}$) — расстояние между вершинами.

Нормированная центральность по близости:

$$
C_C'(i) = \frac{g - 1}{\sum_{j \neq i} d_{ij}}
$$



### 3. Центральность по посредничеству

$$
C_B(i) = \sum_{j < k} \frac{\sigma_{jk}(i)}{\sigma_{jk}}
$$

где:  
- $(\sigma_{jk}$) — число кратчайших путей между $(j$) и $(k$)  
- $(\sigma_{jk}(i)$) — число таких путей, проходящих через $(i$)

Нормированная центральность:

$$
C_B'(i) = \frac{2 C_B(i)}{(g - 1)(g - 2)}
$$



### 4. Центральность по информации

Алгоритм вычисления:

1. Матрица псевдосмежности:
   $
   A = I + K
   $
   где $K$ — матрица Кирхгофа.

2. Найти обратную матрицу:
   $
   C = A^{-1}
   $

3. Вычислить:
   $
   T = \mathrm{Tr}(C), \qquad R = \sum_{i,j} c_{ij}
   $

Центральность по информации:

$$
C_I(i) = \frac{1}{c_{ii} + \frac{T - 2R}{g}}
$$

Нормированная центральность:

$$
C_I'(i) = \frac{C_I(i)}{\sum_j C_I(j)}
$$


## Групповые характеристики сетей



### 1. Централизация по степени

Групповая централизация отражает, насколько распределение степеней отклоняется от максимально возможного (например, звезды).

$$
C_D = \frac{\sum_{i=1}^{g} \bigl(C_D(n^*) - C_D(n_i)\bigr)}{(g - 1)(g - 2)}
$$

где $(C_D(n^*)$) — значение максимальной степени центральности в сети; $(g$) — число вершин.



### 2. Централизация по близости

$$
C_C = \frac{\sum_{i=1}^{g} \bigl(C'_C(n^*) - C'_C(n_i)\bigr)}{\dfrac{(g - 2)(g - 1)}{2g - 3}}
$$

где $(C'_C$) — нормированные значения центральности по близости, а $(C'_C(n^*)$) — их максимальное значение в сети.



### 3. Централизация по посредничеству

$$
C_B = \frac{\sum_{i=1}^{g} \bigl(C'_B(n^*) - C'_B(n_i)\bigr)}{g - 1}
$$

где $(C'_B$) — нормированная центральность по посредничеству, а $(C'_B(n^*)$) — её максимальное значение в сети.

### Меры престижа в ориентированных сетях

#### 1. Престиж по степени
$$
P_D(i) = \frac{deg_i(i)}{\,g - 1\,}
$$

где $deg_i(i)$ — входящая полустепень вершины $i$.



#### 2. Престиж по близости
$$
P_P(i) = \frac{|I_i|}{\,g-1\,} \cdot \frac{|I_i|}{\sum_j d_{ij}}
$$

где $|I_i|$ — количество вершин, достижимых из вершины $i$.



#### 3. Престиж по рангу

Для вычисления рангового престижа решается система:

$$
\left(\frac{1}{a}E - X\right)^T P_R = x
$$

где:  
- $X$ — матрица смежности (социоматрица),  
- $a = \frac{1}{\lambda}$, где $\lambda$ — максимальное собственное значение матрицы $X$,  
- $x$ — вектор сумм столбцов,  
- $E$ — единичная матрица.  

Нормированный престиж:

$$
P_R'(n_i) = \frac{P_R(n_i)}{m}
$$

где  

$$
m = \sum_{k=1}^{g-1} a^k A_{g-1}^k .
$$


### Алгоритмы визуализации сетей



#### Алгоритм Фрюхтермана-Рейнгольда

Основан на физической аналогии:  
- Вершины отталкиваются друг от друга, подобно зарядам  
- Рёбра ведут себя как пружины, соединяющие вершины  

Цель — минимизация функции потенциальной энергии системы:

$$
E = \sum_{i \neq j} k_r(d_{ij}) + \sum_{(i,j) \in E} k_a(d_{ij})
$$

где $k_r$ — коэффициент силы отталкивания,  
$k_a$ — коэффициент силы притяжения,  
$d_{ij}$ — расстояние между вершинами $i$ и $j$.



#### Алгоритм Камада-Каваи

Минимизирует расхождение между геодезическими расстояниями в графе и евклидовыми расстояниями на плоскости:

$$
E = \sum_{i \neq j} \frac{1}{2} k_{ij} \left(||p_i - p_j|| - l_{ij}\right)^2
$$

где:  
- $p_i, p_j$ — координаты вершин $i$ и $j$,  
- $l_{ij}$ — желаемое расстояние между вершинами,  
- $k_{ij}$ — весовая константа, отражающая важность сохранения расстояния.

### Работа с florentine_business.csv
#### 1. Извлечение из florentine_business.csv социоматрицы

```r
file_url <- "https://drive.google.com/file/d/1shboWsyEAABA1P2EM4CtN5Y6FeRiwZi3/view?usp=sharing"
download_url <- paste0("https://drive.google.com/uc?id=",
                       gsub(".*/d/([^/]+).*", "\\1", file_url))

raw_data <- read.csv(download_url,
                     header = FALSE,
                     skip = 1,
                     sep = ",",
                     stringsAsFactors = FALSE)

family_names <- as.character(raw_data[1, ])
sociomatrix  <- as.matrix(raw_data[-1, ])
mode(sociomatrix) <- "numeric"

if (any(is.na(sociomatrix))) {
  stop("Обнаружены нечисловые значения в социоматрице!")
}

rownames(sociomatrix) <- family_names
colnames(sociomatrix) <- family_names

print(sociomatrix)

florentine_graph <- graph_from_adjacency_matrix(sociomatrix, mode = "undirected")

vertex_size <- igraph::degree(florentine_graph) * 5 + 10

plot(florentine_graph,
     layout = layout_with_fr,
     main = "Флорентийские семьи: Fruchterman-Reingold",
     vertex.size = vertex_size,
     vertex.color = "lightblue",
     vertex.label = V(florentine_graph)$name,
     vertex.label.cex = 0.8,
     vertex.label.color = "black",
     edge.color = "gray50")
```

#### 2. Поиск точек сочленения и мостов

```r

net1 <- network(sociomatrix, matrix.type = "adjacency", directed = FALSE)

# 3. Точки сочленения (cutpoints)
# Поиск точек сочленения (cutpoints)
# Возвращает логический вектор для вершин: TRUE если точка сочленения
cut_vertices <- cutpoints(net1, mode = "graph", return.indicator = TRUE)

# Визуализация: точки сочленения красные
gplot(net1, gmode = "graph",
      vertex.col = ifelse(cut_vertices, "red", "lightblue"),
      vertex.size = 6,
      label = family_names,
      label.cex = 0.9,
      displaylabels = TRUE,
      main = "Точки сочленения (cutpoints)")
#4. Поиск мостов (bridges)
find_bridges <- function(net, mode = "graph", connected = c("strong", "weak")) {
  edge_count <- network.edgecount(net)
  bridge_vec <- rep(FALSE, edge_count)

  # Количество компонент связности исходного графа
  if (mode == "graph") {
    cmp_count <- components(net)
    for (i in 1:edge_count) {
      tmp_net <- net
      delete.edges(tmp_net, i)
      bridge_vec[i] <- (components(tmp_net) != cmp_count)
    }
  } else {
    cmp_count <- components(net, connected = connected)
    for (i in 1:edge_count) {
      tmp_net <- net
      delete.edges(tmp_net, i)
      bridge_vec[i] <- (components(tmp_net, connected = connected) != cmp_count)
    }
  }
  return(bridge_vec)
}

# Получаем логический вектор мостов
bridge_edges <- find_bridges(net1)

#5. Визуализация мостов =
gplot(net1, gmode = "graph",
      vertex.col = "lightblue",
      vertex.size = 6,
      label = family_names,
      label.cex = 0.9,
      displaylabels = TRUE,
      edge.col = ifelse(bridge_edges, "red", "gray70"),   # мосты — красные
      edge.lwd = ifelse(bridge_edges, 3, 1),              # мосты — толще
      main = "Мосты (bridges)")
```

#### 3.  Значение централизации для каждой меры

```r
non_isolated <- which(degree(net1, gmode = "graph") > 0)
net_connected <- net1 %v% "vertex.names"  # просто копируем на всякий случай
if (length(non_isolated) < network.size(net1)) {
  net_connected <- get.inducedSubgraph(net1, non_isolated)
}

# Основные характеристики подграфа без изолятов
g <- network.size(net_connected)               # число вершин
deg <- degree(net_connected, gmode = "graph")  # степени вершин

# Меры центральности
closeness_cent <- closeness(net_connected, gmode = "graph")       # близость
between_cent   <- betweenness(net_connected, gmode = "graph")     # посредничество

# Вывод результатов
cat("Централизации для каждой меры (сеть флорентийских семей):\n\n")

cat("Централизация по степени:\n")
C_D <- sum(max(deg) - deg) / ((g - 1) * (g - 2))
cat("C_D =", round(C_D, 4), "\n\n")

cat("Централизация по близости:\n")
# Стандартная формула для ненаправленного графа
C_C <- sum(max(closeness_cent, na.rm = TRUE) - closeness_cent, na.rm = TRUE) *
  (g - 1) / ((g - 1) * (g - 2) / (2 * g - 3))
cat("C_C =", round(C_C, 4), "\n\n")

cat("Централизация по посреднечиству:\n")
C_B <- sum(max(between_cent) - between_cent) / (((g - 1) * (g - 2)) / 2)
cat("C_B =", round(C_B, 4), "\n")
```

## Вывод:
- начение централизации по степени (C_D ≈ 0.278) указывает на то, что в сети есть отдельные узлы с большим числом связей, но сеть в целом не сильно сосредоточена вокруг одного актора.

- Высокое значение централизации по близости (C_C ≈ 3.535) говорит о том, что расстояния между узлами существенно различаются, и некоторые узлы находятся ближе к остальным, играя роль «центров доступа» в сети.

- Централизация по посредничеству (C_B ≈ 4.422) показывает, что есть узлы, через которые проходит значительная часть кратчайших путей, то есть они обладают большим влиянием на передачу информации внутри сети.

### Работа с сетью аффелированости ceo.csv

#### 6. Извлечение из ceo.csv социоматрицы

```r
url_ceo <- "https://drive.google.com/file/d/17YTBt-r4hE1eoXkVWGY-DIiY1jU1f7uQ/view?usp=drive_link"

download_url_ceo <- paste0("https://drive.google.com/uc?id=",
                           gsub(".*/d/([^/]+).*", "\\1", url_ceo))

raw_ceo <- read.csv(download_url_ceo,
                    header = FALSE,
                    skip = 9,
                    sep = ",",
                    stringsAsFactors = FALSE)

# Преобразуем в числовую матрицу инцидентности (директора × компании)
sociomatrix_ceo <- as.matrix(raw_ceo)
mode(sociomatrix_ceo) <- "numeric"

# Убираем возможные NA, если они есть
sociomatrix_ceo[is.na(sociomatrix_ceo)] <- 0

print("Размер двухмодовой матрицы (директора × компании):")
print(dim(sociomatrix_ceo))

inet2 <- graph_from_incidence_matrix(sociomatrix_ceo,
                                    directed = FALSE)


V(inet2)$type <- bipartite_mapping(inet2)$type

V(inet2)$shape <- ifelse(V(inet2)$type, "circle", "square")   # люди — кружки, компании — квадраты
V(inet2)$color <- ifelse(V(inet2)$type, "lightblue", "orange")
V(inet2)$label <- ifelse(V(inet2)$type,
                        paste("D", 1:sum(V(inet2)$type), sep=""),
                        paste("C", 1:sum(!V(inet2)$type), sep=""))
V(inet2)$label.cex <- 0.7
V(inet2)$size <- ifelse(V(inet2)$type, 12, 18)

plot(inet2,
     layout = layout_with_kk,
     main = "Двухмодовая сеть «Директора — Компании»\n(кружки = директора, квадраты = компании)",
     vertex.frame.color = "white",
     edge.color = "gray70",
     edge.width = 1.5)
```

#### 7. Отображение сети в виде биграфа

```r
plot(inet2,
     layout = layout_as_bipartite ,
     main = "Двухмодовая сеть «Директора — Компании»\n(кружки = директора, квадраты = компании)",
     vertex.frame.color = "white",
     edge.color = "gray70",
     edge.width = 1.5)
```

#### 8. Отображение сети в виде гиперграфа

```r
types <- V(inet2)$type  # TRUE = один мод (клубы/компании), FALSE = другой (CEO/директора)


ceo_ids   <- which(!types)   # директора
club_ids  <- which(types)    # компании (клубы)

# Получаем список смежности
adj_list <- as_adj_list(inet2)

# Создаём гиперрёбра: каждое гиперребро = все директора, входящие в одну компанию
hyperedges <- list()

for (club in club_ids) {
  # Соседи клуба = все директора, которые в него входят
  members <- neighbors(inet2, club)

  # Оставляем только тех, кто действительно директор
  members <- members[members %in% ceo_ids]

  if (length(members) >= 2) {  # гиперребро имеет смысл только если ≥2 участника
    hyperedges[[paste0("Club_", club)]] <- as.character(members)
  }
}


hnet2 <- hypergraph_from_edgelist(hyperedges)

plot(inet2,
     layout = layout_with_fr,
     vertex.color = ifelse(types, "salmon", "lightblue"),
     vertex.shape = ifelse(types, "square", "circle"),
     vertex.size = ifelse(types, 20, 10),
     vertex.label = ifelse(types, paste("C", 1:length(club_ids), sep=""), ""),
     vertex.label.cex = 0.9,
     mark.groups = lapply(club_ids, function(x) neighbors(inet2, x)),
     mark.col = rainbow(length(club_ids), alpha = 0.15),
     mark.border = rainbow(length(club_ids), alpha = 0.6),
     main = "Двухмодовая сеть как гиперграф\n(цветные области = состав одной компании)")
```

#### 9. Представление сети аффилированности в виде двух одномодальных сетей

```r
proj <- bipartite_projection(inet2, multiplicity = TRUE)

ceo_net  <- proj$proj1    # сеть директоров: ребро = сколько общих компаний
club_net <- proj$proj2    # сеть компаний: ребро = сколько общих директоров

cat("Матрица совместного участия директоров (вес = кол-во общих компаний):\n")
print(as_adjacency_matrix(ceo_net, attr = "weight", sparse = FALSE))

cat("\nМатрица пересечения компаний (вес = кол-во общих директоров):\n")
print(as_adjacency_matrix(club_net, attr = "weight", sparse = FALSE))


plot(ceo_net,
     layout = layout_with_fr,
     vertex.color = "lightblue",
     vertex.shape = "circle",
     vertex.size = 15,
     vertex.label = paste0("CEO", 1:vcount(ceo_net)),
     vertex.label.cex = 0.8,
     vertex.label.color = "black",
     edge.width = E(ceo_net)$weight * 1.5,
     edge.color = "gray50",
     main = "Сеть директоров\n(толщина ребра = число общих компаний)")

plot(club_net,
     layout = layout_with_fr,
     vertex.color = "salmon",
     vertex.shape = "square",
     vertex.size = 20,
     vertex.label = paste0("Comp", 1:vcount(club_net)),
     vertex.label.cex = 0.9,
     edge.width = E(club_net)$weight * 1.5,
     edge.color = "gray40",
     main = "Сеть компаний\n(толщина ребра = число общих директоров)")
```

##### Вывод
Анализ перекрытий в советах директоров компаний позволяет выявить поляризованную структуру связей, характерную для сетей аффилиации. Наиболее заметной чертой является формирование насыщенного ядра — небольшой группы лиц, которые одновременно занимают руководящие позиции в нескольких организациях. Этот кластер сконцентрированных связей визуально выделяется на графе за счёт плотного пучка рёбер и подтверждается данными матрицы смежности. Данный паттерн отражает механизм «замкнутого круга» в корпоративном управлении, когда ограниченная группа директоров многократно взаимодействует через совместное членство в ключевых компаниях, усиливая взаимное влияние и создавая устойчивые коалиции.

На периферии сети, напротив, расположены акторы с единичными или слабыми связями, что свидетельствует об их фрагментарном вовлечении в структуры корпоративного управления. Такие участники обычно представлены в советах лишь одной или двух компаний и не образуют устойчивых перекрытий с другими директорами. Таким образом, сеть перекрытий директоров формирует чёткую иерархию: в центре находится сплочённая элитная группа, осуществляющая стратегическое влияние через перекрёстное членство, а на окраинах — слабо интегрированные участники, выполняющие преимущественно символические или номинальные роли. Такая конфигурация соответствует классической модели централизованной корпоративной элиты, где власть и ресурсы сосредоточены в руках ограниченного круга лиц.

#### 10. Вычисление степени участия и среднее количество событий для каждого актора

```r

# Степень участия каждого директора = сколько компаний в которых он состоит
participation <- rowSums(sociomatrix_ceo)

# Среднее количество компаний на одного директора
mean_participation <- mean(participation)

cat("Степень участия директоров (сколько компаний у каждого):\n")
print(sort(participation, decreasing = TRUE))

cat("\nСреднее количество компаний на одного директора:",
    round(mean_participation, 3), "\n")

cat("Медиана:", median(participation), "\n")
cat("Минимум:", min(participation), "\n")
cat("Максимум:", max(participation), "\n")

company_members <- colSums(sociomatrix_ceo)

cat("\nСреднее количество директоров в одной компании:",
    round(mean(company_members), 3), "\n")
cat("Самая большая компания имеет", max(company_members), "директоров\n")
```

##### Вывод
Аффилиационная сеть директоров характеризуется выраженным неравенством в степени участия: небольшая группа высокоактивных директоров формирует плотное ядро перекрестных связей, в то время как остальные распределены по уровням вовлечённости от умеренного до минимального. Такая структура типична для сетей корпоративного управления, где ресурсы и влияние концентрируются в руках ограниченного круга лиц.

#### 14. Определение логарифма отношения шансов.


```r
n_clubs <- ncol(sociomatrix_ceo)

# Создаём пустую матрицу (симметричную)
logOR <- matrix(0, nrow = n_clubs, ncol = n_clubs)
rownames(logOR) <- colnames(logOR) <- paste0("Comp", 1:n_clubs)

for (i in 1:(n_clubs - 1)) {
  for (j in (i + 1):n_clubs) {
    a <- sum(sociomatrix_ceo[, i] == 1 & sociomatrix_ceo[, j] == 1) + 0.5   # оба
    b <- sum(sociomatrix_ceo[, i] == 1 & sociomatrix_ceo[, j] == 0) + 0.5   # только i
    c <- sum(sociomatrix_ceo[, i] == 0 & sociomatrix_ceo[, j] == 1) + 0.5   # только j
    d <- sum(sociomatrix_ceo[, i] == 0 & sociomatrix_ceo[, j] == 0) + 0.5   # ни один

    odds_ratio <- (a * d) / (b * c)
    log_or_val <- round(log(odds_ratio), 3)

    logOR[i, j] <- logOR[j, i] <- log_or_val
  }
}

cat("Матрица логарифма отношения шансов между компаниями:\n")
cat("(>0 — чаще встречаются вместе, чем случайно; <0 — реже)\n")
print(round(logOR, 3))
```

##### Вывод
На основе анализа матрицы отношения шансов выявлена структурированная сеть перекрестного директорства. Образовались кластеры тесно связанных компаний: Comp1, Comp4 и Comp10 демонстрируют сильные положительные связи (значения до 2.37), что указывает на частые случаи, когда одни и те же директора входят в советы этих компаний, формируя коалицию. Comp2, Comp3, Comp6, Comp11, Comp12, Comp13 и Comp15 также образуют сплочённую группу. При этом существуют и пары с устойчиво отрицательной связью (например, Comp3 и Comp4: -2.73), где совместное членство директоров встречается реже ожидаемого, вероятно из-за конкуренции или регуляторных барьеров. Остальные пары демонстрируют нейтральные или случайные связи. Таким образом, сеть неоднородна и разделена на блоки с высокой и низкой степенью перекрытия директоратов.
