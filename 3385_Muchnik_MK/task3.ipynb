## Структурные показатели

**$k$-ядро** — подграф, в котором каждый узел имеет не менее $k$ связей с другими узлами внутри него.

---

## Анализ диад

Диада — пара акторов и связей между ними. Типы диад:
- **Взаимные** ($M$): связь в обоих направлениях
- **Асимметричные** ($A$): связь только в одну сторону
- **Нулевые** ($N$): связей нет

$$
M + A + N = g(g-1)
$$

**Цензус диад** — тройка $(M, A, N)$, описывающая структуру направленных связей в сети.

---

## Взаимность в сетях

**Индекс Каца–Пауэлла** оценивает склонность к взаимным связям:

$$
\mu_{\text{кр}} = \frac{2(g - 1)M - gd^2}{gd(g - d - 1)}
$$

где $d$ — число возможных выборов каждого актора.

**Индекс Бондопаддхая** сравнивает наблюдаемую взаимность с теоретическими пределами:

$$
M_B = \frac{M - M_{\min}}{M_{\max} - M_{\min}}
$$

---

## Централизация сети

Показывает, насколько сеть сосредоточена вокруг одного или нескольких ключевых акторов.

**По степени** (количеству связей):
$$
C_D = \frac{\sum_{i=1}^g [C_D(n^*) - C_D(n_i)]}{(g-1)(g-2)}
$$

**По близости** (среднему расстоянию до других):
$$
C_C = \frac{\sum_{i=1}^g [C'_C(n^*) - C'_C(n_i)]}{[(g-2)(g-1)]/(2g-3)}
$$

**По посредничеству** (контролю над потоками):
$$
C_B = \frac{2 \sum_{i=1}^g [C'_B(n^*) - C'_B(n_i)]}{(g-2)(g-1)^2}
$$

где $n^*$ — наиболее центральный актор.


**Условная вероятность взаимности:**

$$
P(j \rightarrow i \mid i \rightarrow j) = P(j \rightarrow i) + \rho_{kp} P(j \not\rightarrow i)
$$

где $\rho_{kp}$ — параметр взаимности:
- $\rho_{kp} = 0$: выборы независимы (нет взаимности)
- $\rho_{kp} = 1$: полная взаимность (выбор $i \rightarrow j$ влечёт $j \rightarrow i$)

**Ожидаемое число взаимных диад с учётом $\rho_{kp}$:**

$$
E(M \mid \rho_{kp}) = \frac{g d^2}{2(g-1)} (1 - \rho_{kp}) + \frac{g d}{2} \rho_{kp}
$$

**Оценка параметра взаимности (метод моментов):**

$$
\hat{\rho}_{kp} = \frac{2(g-1)M - g d^2}{g d (g-1-d)}
$$

где $M$ — наблюдаемое число взаимных диад.

**Индекс взаимности Каца-Пауэлла для свободного выбора**

Пусть $x_{i+}$ — исходящая степень $i$-го актора,
$L = \sum x_{i+}$ — общее число дуг,
$L_2 = \sum x_{i+}^2$ — сумма квадратов степеней.

**Ожидаемое число взаимных диад при случайном выборе:**

$$
E(M) = \frac{L^2 - L_2}{2(g-1)^2}
$$

**Ожидаемое число взаимных диад с учётом $\rho'_{kp}$:**

$$
E(M \mid \rho'_{kp}) = \frac{L^2 - L_2}{2(g-1)^2} (1 - \rho'_{kp}) + \frac{L}{2} \rho'_{kp}
$$

**Оценка параметра взаимности (метод моментов):**

$$
\hat{\rho}'_{pk} = \frac{2(g-1)^2 M - L^2 + L_2}{L(g-1)^2 - L^2 + L_2}
$$

**Индекс Бондопаддхая:**

Пусть $M_{min}$ и $M_{max}$ — минимально и максимально возможное
число взаимных диад при заданных степенях вершин.

**Вычисление $M_{min}$ и $M_{max}$:**

Для $t = 0, 1, \ldots, g$ определим:

$$
f(t) = \sum_{i=1}^{t} x_{i+} - t(g-1) - \binom{t}{2} \tag{13.16}
$$

$$
g(t) = \sum_{i=1}^{t} x_{i+} - t(t-1) - \sum_{i=t+1}^{g} \min(t, x_{i+}) \tag{13.17}
$$

Тогда:

$$
M_{min} = \max_{0 \leq t \leq g} f(t)
$$

$$
M_{max} = \left\lfloor \frac{\sum_{i=1}^{g} x_{i+} - \max_{0 \leq t \leq g} g(t)}{2} \right\rfloor

### Диады и цензус диад

#### Диада
Диада (dyad) — неупорядоченная пара акторов и направленных ребер, которые существуют между ними
$$ D_{ij} = (x_{ij}, x_{ji})$$

- симметричная (mutual): дуги $  i \to j $ и $ j \to i $ присутствуют одновременно - (1,1)

$$ M = 1/2 * tr(X^2)$$

- ассиметричная (asymmetric): присутствует ровно одна из двух возможных дуг - (1,0) или (0,1)

$$ A = tr(XX^T) - 2M$$

- нулевая (null): обе дуги отсутствуют - (0,0)

$$ N = C^2_g - A - M $$

#### Цензус диад
Цензус диад (dyad census) — тройка объектов $ (M, A, N) $

- $M+A+N = g(g-1)$

### Индекс взаимности Каца–Пауэлла

Характеристика измерения тенденции акторов в группе
$$
\mu_{\text{кр}} = \frac{2(g - 1)M - gd^2}{gd(g - d -1)},
$$

где $d ≤ g - 1$ - число выборов, который может сделать каждый актор

- фиксированная модель: предполагается, что каждый узел имеет одинаковую ожидаемую степень $ d $
- свободная модель: учитывает реальное распределение входящих и исходящих степеней

#### Вероятность взаимной связи

В рамках модели Каца–Пауэлла оценивается вероятность того, что диада $ ij $ сопровождается диадой $ ji $. Вычисляется отдельно для:
- фиксированного выбора
- свободного выбора

### Индекс взаимности Бондопаддхая

Индекс, основанный на минимальном и максимальном значении количества взаимных диад:
$$
M_B = \frac{M - M_{\min}}{M_{\max} - M_{\min}},
$$
где:
- $ M $ — наблюдаемое число взаимных диад
- $ M_{\min} ,  M_{\max} $ — минимально и максимально возможное число взаимных диад при данном распределении степеней

## Выполнение работы

###1. Моделирование случайного орграфа

```r
if (!require(statnet)) install.packages("statnet", repos = "https://cloud.r-project.org/")
if (!require(igraph)) install.packages("igraph", repos = "https://cloud.r-project.org/")

library(statnet)
library(igraph)
```

```r
g <- 111
p <- 0.62
graph <- erdos.renyi.game(n = g, p = p, directed = TRUE)
```

###2. Отображение полученного орграфа

```r
plot(graph,
     vertex.size = 8,
     vertex.color = "lightblue",
     edge.arrow.size = 0.1,
     edge.arrow.width = 0.3,
     edge.color = adjustcolor("gray", alpha.f = 0.3),
     layout = layout_with_fr)
```

###3. Вывод "пятичисловой сводки"

```r
adj_matrix <- as_adjacency_matrix(graph, sparse = FALSE)
net <- network(adj_matrix, matrix.type = "adjacency", directed = TRUE)

g1 <- network.size(net)
cat("Размер графа:", g1, '\n')

cat("Плотность графа:", gden(net), '\n')

# количество компонент связности
comp <- components(graph, mode = "strong")
cat("Число компонент связности:", comp$no, '\n')

# диаметр сети
lgc1 <- largest_component(graph, mode = "strong")
cat("Диаметр графа:", diameter(lgc1, directed = TRUE), '\n')

# коэффициент кластеризации
cat("Коэффициент кластеризации", gtrans(net, mode = "disgraph"))
```

###4. Вычисление цензуса диад

```r
d_c <- dyad_census(graph)
d_c
```

Выводы:

* 2383 — взаимные связи (A <-> B)
* 2874 — асимметричные связи (A -> B, но не B -> A)
* 848  — нулевые связи (A и B не связаны)
* Есть заметная асимметрия

Плотность графа 0.62, поэтому асимметричных диад больше остальных

###5. Вычислить оценку индекса взаимности Каца—Пауэла при фиксированом выборе акторов 𝑑 и при свободном выборе

```r
d <- (g-1)*p

# фиксированный выбор
fix <- (2*(g - 1)*d_c$mut - g*d*d)/(g*d*(g - d - 1))
cat("Индекс взаимности КП с фиксированным d =", d, "\n", fix)

cat("\n\n")

# свободный выбор
degree_dist <- degree(graph)
degree_list_out <- degree(graph, mode = "out")
L <- sum(degree_list_out)
L2 <- sum(degree_list_out^2)
no_fix <- (2*(g - 1)*(g - 1)*d_c$mut - L*L + L2)/(L*(g - 1)*(g - 1) - L*L + L2)
cat("Оценка индекса взаимности КП при свободном выборе:\n", no_fix)
```

Выводы:

Сеть имеет отрицательную взаимность по Кацу, что свидетельствует о умеренной взаимности

###6. Вычислить вероятность того, что связь будет взаимной для каждого случая

```r
mut_fix <- (d/(g - 1))*((d/(g - 1)) + fix*(g - d - 1)/(g - 1))
cat("Вероятность взаимной связи с фикс d:\n", mut_fix, "\n\n")

mut_no_fix <- (L/(g*(g - 1)))*((L/(g*(g - 1)))  + no_fix*(1 - (L/(g*(g - 1)))))
cat("Вероятность взаимной связи свободный выбор:\n", mut_no_fix)
```

###7. Вычислить индекс взаимности Бондопаддхая

```r
max_g <- -Inf
M_min <- -Inf

for (t in 1:g) {
    f_t <- sum(degree_list_out[1:t]) - t*(g - 1) - (t*(t - 1))/2
    if (t == g)
        sum_min_rest <- 0
    else
        sum_min_rest <- sum(pmin(t, degree_list_out[(t + 1):g]))
    g_t <- sum(degree_list_out[1:t]) - t*(t - 1) - sum_min_rest
    if (g_t > max_g) max_g <- g_t
    if (f_t > M_min) M_min <- f_t
}

M_min <- pmax(0, M_min)
M_max <- (sum(degree_list_out) - max_g) / 2

cat("M_max: ", M_max, '\n')
cat("M_min: ", M_min, '\n')
M_b <- (d_c$mut - M_min)/(M_max - M_min)
cat("Индекс взаимности Бондопаддхая: ", M_b)
```

Выводы:

Сеть умеренно взаимна (взаимность больше средней)

###8. Проверить нулевую гипотезу. Определить максимальный уровень значимости при котором нулевая гипотеза не отвергается ($1/2$, $\alpha$=0,10)

```r
n <- g*(g-1)
P0 <- 0.73
E_L <- n * P0
Var_L <- n * P0 * (1 - P0)
sd_L <- sqrt(Var_L)

z <- (L - E_L) / sd_L
p_value <- 2 * pnorm(-abs(z))
alpha <- 0.05
reject_H0 <- abs(z) > qnorm(1 - alpha/2)

L
round(z, 4)
format.pval(p_value)
cat("Нулевая гипотеза", ifelse(reject_H0, "отвергается", "не отвергается"),
      "на уровне α =", alpha, "\n")
p_value
```

###9. Проверить нулевую гипотезу. Определить максимальный уровень значимости при котором нулевая гипотеза не отвергается ($P_0 = 0,73$)

```r
n <- g*(g-1)
E_L_H0 <- n * P0
Var_L_H0 <- n * P0 * (1 - P0)
sd_L_H0 <- sqrt(Var_L_H0)
z <- (L - E_L_H0) / sd_L_H0
p_value <- 2 * pnorm(-abs(z))
c_value <- qnorm(1 - alpha/2)
reject_H0 <- abs(z) > c_value
cat("Нулевая гипотеза", ifelse(reject_H0, "отвергается", "не отвергается"),
      "на уровне α =", alpha, "\n")
p_value
```

###10. Построить доверительный интервал для оценки параметра 𝑝 с доверительной надёжностью 1 − $\alpha$

```r
p_hat <- L/n
se <- sqrt(p_hat*(1 - p_hat)/n)
z <- qnorm(1 - alpha/2)
ci_low <- max(0, p_hat - z*se)
ci_up  <- min(1, p_hat + z*se)

cat("p:", p_hat, "\n")
cat(sprintf("%.0f%% доверительный интервал для p: [%.4f, %.4f]\n", (1 - alpha) * 100, ci_low, ci_up))
cat("p = 0.58", ifelse(ci_low <= 0.58 && 0.58 <= ci_up, "попадает", "не попадает"), "в доверительный интервал")

```

Выводы:

95% доверительный интервал [0.6171, 0.6343]достаточно узкий

Реальное значение p = 0.58 лежит внутри этого интервала
